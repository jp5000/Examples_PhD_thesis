# 3) Multivariate pdf
## Only one output or quantity of interest (QoI)

OAKLEY & O'HAGAN (2004) FUNCTION

$y(\mathbf{x}) = \mathbf{a_1}^T \mathbf{x} + \mathbf{a_2}^T \sin(\mathbf{x}) + \mathbf{a_3}^T \cos(\mathbf{x}) + \mathbf{x}^T \mathbf{M} \mathbf{x}$

Dimensions: 15

Input Distributions PDF$(\mathbf{x})$:

The independent distributions of the input random variables are: xi ~ N(μ=0, σ=1), for all i = 1, …, 15. 


Description from https://www.sfu.ca/~ssurjano/oakoh04.html

This function's a-coefficients are chosen so that 5 of the input variables contribute significantly to the output variance, 5 have a much smaller effect, and the remaining 5 have almost no effect on the output variance. Values of the coefficient vectors a1, a2 and a3, and the matrix M, can be found at:
http://www.jeremy-oakley.staff.shef.ac.uk/psa_example.txt.

Generate a random covariance matrix by generating a symmetric positive definite matrix

````python
C = np.eye(15)
````

````python
dist_X = cp.Iid(cp.Normal(),15)
dist_W = cp.Iid(cp.Normal(),15)
````

### a) MC with Halton sampling

````python
N_MC = int(5e5)
sample_inputs = dist_X.sample(size=N_MC,rule='H')
sample_u = np.array([y(node) for node in sample_inputs.T])
df_MC = pd.DataFrame(columns=['x_'+str(i) for i in range(15)]+['y'],data=np.vstack([sample_inputs,sample_u]).T)

E_u = np.mean(sample_u,axis=0)
Std_u = np.std(sample_u,axis=0)
Q05_u = np.percentile(sample_u,5,axis=0)
Q95_u = np.percentile(sample_u,95,axis=0)
````

### b) PCE with Point collocation

````python
M = 3    # Order of the polynomial
N = M + 1 # Number of terms

P = cp.orth_ttr(M,dist_W)
````

````python
# Halton sequence sample
nodes_W = dist_W.sample(3.*special.comb(M+15,15),rule='H')

print( 'nodes', len(nodes_W[0,:]) )

nodes_X = dist_X.inv(dist_W.fwd(nodes_W))
sample_u_pc = np.array([y(node) for node in nodes_X.T])

df_train=pd.DataFrame(columns=['x_'+str(i) for i in range(15)]+['y'],data=np.vstack([nodes_X,sample_u_pc]).T)
````

````python
# Point collocation
u_hat = cp.fit_regression(P, nodes_W, sample_u_pc, 
                          #rule='LA',n_nonzero_coefs=20) #Lars (Least Angle Regression). n_nonzero_coefs= all dimensions
                          #rule='LAS',alpha=1e-4,max_iter=200) #Lasso (Least Absolute Shrinkage and Selection Operator) 
                          rule='LL', alpha=1e-4,max_iter=200) #Lasso and Lars model
                          #rule='OMP',n_nonzero_coefs=M+1, tol=1e2) # Orthogonal Matching Pursuit. n_nonzero_coefs=1D
                          # -----------
                          # NOT sparse
                          # -----------
                          #rule='BARD')

# Statistics based on the PCE coefficients
mean_u_pc, std_u_pc = cp.E(u_hat, dist_W), cp.Std(u_hat, dist_W)

# Print the fit in ordered way
print( len(u_hat.A) )
for key in u_hat.A:
    print (key , ' ' , u_hat.A[key])
````

# SA

````python
df_sens = pd.DataFrame(columns=['w_'+str(i) for i in range(15)],index=['S_i','S_Ti'],data=0.)
sum_c2 = np.sum(np.array(u_hat.A.values())**2.) - cp.E(u_hat,dist_W)
fact = 0.5*sum_c2/cp.Var(u_hat,dist_W)
for key in u_hat.A:
    if np.sum(np.array(key))>0:
        print (key,' ', np.where(np.array(key)>0)[0] , ' ' , u_hat.A[key]**2./sum_c2)
        ind_Ti = np.where(np.array(key)>0)[0]
        for i in ind_Ti:
            df_sens.iloc[1,i] += fact*u_hat.A[key]**2./sum_c2
        if len(ind_Ti)==1:
            df_sens.iloc[0,ind_Ti[0]] += fact*u_hat.A[key]**2./sum_c2
````
